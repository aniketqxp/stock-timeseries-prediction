```python
import yfinance as yf
import pandas as pd
```

```python
it_stocks = ['INFY.NS', 'WIPRO.NS', 'TCS.NS', 'HCLTECH.NS', 'TECHM.NS']

# Download historical stock data for the selected IT companies without limiting the date range
data = yf.download(it_stocks)

data.columns = [f"{col[1]}_{col[0]}" for col in data.columns]

# Print the first few rows of the data to verify
print(data.tail())

# Calculate and print the total number of observations for each company
print("\nTotal number of observations for each company:")
for ticker in it_stocks:
    # Access Adj Close columns properly by using the format 'ticker_Adj Close'
    adj_close_column = f"{ticker}_Adj Close"
    print(f"{ticker}: {data[adj_close_column].dropna().shape[0]}")
```

```python
print("Column Names:")
print(data.columns)

# Display detailed information about each column/feature
print("\nDetailed Column/Feature Information:")
data.info()
```

```python
data.to_csv('IT_Stocks_Historical_Data.csv')
print("\nData saved to 'IT_Stocks_Historical_Data.csv'")
```

```python
missing_counts = data.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]  # Only show columns with missing values

print("Missing values count (column-wise and company-wise):\n")
for column, count in missing_counts.items():
    print(f"{column}: {count} missing values")
```
